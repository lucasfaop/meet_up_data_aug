```python
from textattack.augmentation.recipes import CheckListAugmenter
augmenter = CheckListAugmenter(transformations_per_example=5)
```

```python
text = "Fred travel to Paris in France with only 2.000 dollars"
```

```python
augmented_texts = augmenter.augment(text)
print("Original:")
print(text)
print("Augmented Texts:")
print(augmented_texts)
```
