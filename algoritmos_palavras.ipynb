Algoritmos Caracteres

```python
import nlpaug.augmenter.word as naw
```

```python
text = "Eu adoro NLP e treinar modelos como BERT!"
```

# Troca por sinônimos

```python
import nltk
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet', 'por')
nltk.download('wordnet')
nltk.download('omw')
```

```python
aug = naw.SynonymAug(aug_src='wordnet', lang='por')
augmented_texts = aug.augment(text, n=3)
print("Original:")
print(text)
print("Augmented Texts:")
print(augmented_texts)
```

# EDA

```python
from textattack.augmentation.recipes import EasyDataAugmenter
augmenter = EasyDataAugmenter(pct_words_to_swap=0.2, transformations_per_example=8)
```

```python
augmented_texts = augmenter.augment(text)
print("Original:")
print(text)
print("Augmented Texts:")
print(augmented_texts)
```

# Embedding Bert

```python
import nlpaug.flow as naf
```

```python
aug_bert = naf.Sequential([naf.Sometimes([
    naw.ContextualWordEmbsAug(model_path='neuralmind/bert-base-portuguese-cased', action="insert", device ='cuda')]),
    naf.Sometimes([naw.ContextualWordEmbsAug(model_path='neuralmind/bert-base-portuguese-cased', action="substitute", device ='cuda')
])])
```

```python
augmented_texts = aug_bert.augment(text, n=3)
print("Original:")
print(text)
print("Augmented Texts:")
print(augmented_texts)
```
